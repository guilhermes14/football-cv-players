```python
#Assessment Vis√£o Computacional
#Bibliotecas utilizadas

import cv2 #OpenCV
import os
import numpy as np 
import time
```

```python
#Captura do video
video = cv2.VideoCapture('napoli_monza.mp4')
framing,image = video.read()
count = 0
framing = True
idx = 0 #index
```

```python
#Leitura de cada frame
while framing:
	
	#Convers√£o para imagem hsv
	hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

    #Cores(rede neural poderia fazer reconhecimento)
	#Verde
	lower_green = np.array([40,40, 40])
	upper_green = np.array([70, 255, 255])
	#Azul
	lower_blue = np.array([110,50,50])
	upper_blue = np.array([130,255,255])
	#Vermelho
	lower_red = np.array([0,31,255])
	upper_red = np.array([176,255,255])
	#Branco
	lower_white = np.array([0,0,0])
	upper_white = np.array([0,0,255])

    #Separa a cor verde do v√≠deo
	mask = cv2.inRange(hsv, lower_green, upper_green)
	res = cv2.bitwise_and(image, image, mask=mask)

	#Convers√£o de hsv para cinza
	res_gray = cv2.cvtColor(res,cv2.COLOR_BGR2GRAY)

    #Transforma√ß√£o da morfologia
	kernel = np.ones((13,13),np.uint8)
	thresh = cv2.threshold(res_gray,127,255,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]#binarization 
	thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
	

    #Procura contornos das morfologias    
	contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
	prev = 0
	font = cv2.FONT_HERSHEY_SIMPLEX
	
	for c in contours:
		x,y,w,h = cv2.boundingRect(c) #Desenho do ret√¢ngulo
		
		#Procura os jogadores e filtra
		if(h>=(1.5)*w):
			if(w>15 and h>= 15):

				idx = idx+1
				player_img = image[y:y+h,x:x+w]
				player_hsv = cv2.cvtColor(player_img,cv2.COLOR_BGR2HSV)

				#Jogadores do Napoli - Azul
				mask1 = cv2.inRange(player_hsv, lower_blue, upper_blue)
				res1 = cv2.bitwise_and(player_img, player_img, mask=mask1)
				res1 = cv2.cvtColor(res1,cv2.COLOR_HSV2BGR)
				res1 = cv2.cvtColor(res1,cv2.COLOR_BGR2GRAY)
				nzCount = cv2.countNonZero(res1)

				#jogadores do Monza - Vermelho
				mask2 = cv2.inRange(player_hsv, lower_red, upper_red)
				res2 = cv2.bitwise_and(player_img, player_img, mask=mask2)
				res2 = cv2.cvtColor(res2,cv2.COLOR_HSV2BGR)
				res2 = cv2.cvtColor(res2,cv2.COLOR_BGR2GRAY)
				nzCountred = cv2.countNonZero(res2)

				if(nzCount >= 20):
					#Napoli
					cv2.putText(image, 'Napoli', (x-2, y-2), font, 0.8, (255,0,0), 2, cv2.LINE_AA)
					cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3)
				else:
					pass

				if(nzCountred>=20):
					#Monza
					cv2.putText(image, 'Monza', (x-2, y-2), font, 0.8, (0,0,255), 2, cv2.LINE_AA)
					cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),3)
				else:
					pass	
				
	time.sleep(0.001)
	print(('Read a new frame: '), framing)     	
	count += 1

	#Janela da aplica√ß√£o
	cv2.imshow('Napoli-Monza',image)

	#cv2.imshow('Napoli-Monza',hsv) #Efeito da convers√£o para HSV
	#cv2.imshow('Napoli-Monza',res) #Filtro reconhecendo a cor verde apenas
	#cv2.imshow('Napoli-Monza',mask) #Filtro da m√°scara 
	#cv2.imshow('Napoli-Monza',thresh) #Filtro para contornos


	if cv2.waitKey(1) & 0xFF == ord('q'): 
		break
	framing,image = video.read()
    
video.release()
cv2.destroyAllWindows()#Desliga o sistema
```
